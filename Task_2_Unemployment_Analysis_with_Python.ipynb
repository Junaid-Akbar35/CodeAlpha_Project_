### **Task 2: Unemployment Analysis with Python**

This task involves analyzing the unemployment rate data, especially focusing on the effects of COVID-19 on unemployment trends. We will clean, explore, and visualize the data to derive meaningful insights.

### **Dataset Overview**
- **Source**: [Kaggle - Unemployment in India](https://www.kaggle.com/datasets/gokulrajkmv/unemployment-in-india).
- **Features**:

  - **Date**: The timeframe of the unemployment rate record.

### **Analysis Objectives**
1. Understand trends in the unemployment rate across time and regions.
2. Examine the impact of COVID-19 on unemployment rates.
3. Visualize key patterns and differences.

### **Complete Code for Task 2**

#### **Step 1: Import Libraries**


```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### **Step 2: Load the Dataset**

```python
# Load the dataset
url = "/content/Unemployment in India.csv"
df = pd.read_csv(url)

# Display the first few rows
print("First 5 Rows of the Dataset:")
print(df.head())

# Check dataset structure
print("\nDataset Info:")
df.info()
```

### **Step 3: Explore and Clean the Data**

```python
# Display the first few rows
print("First 5 Rows of the Dataset:")
print(df.head())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Fill or drop missing values if necessary
df = df.dropna()

# Print the columns of the DataFrame to verify the actual column name
print(df.columns)

# Convert 'Date' column to datetime, adjusting the column name if needed
# Assuming the actual column name is ' Date' based on the common issue of extra space
df[' Date'] = pd.to_datetime(df[' Date'])

# Check unique regions
print("\nUnique Regions:")
print(df['Region'].unique())
```

### **Step 4: Analyze the Data**
This part of the code focuses on extracting insights from the unemployment data, particularly looking at regional differences and the impact of COVID-19.

```python
# prompt: Step 4: Analyze the Data

# Calculate the average unemployment rate for each region
region_avg_unemployment = df.groupby('Region')[' Estimated Unemployment Rate (%)'].mean()
print("\nAverage Unemployment Rate by Region:\n", region_avg_unemployment)

# Analyze the impact of COVID-19 (assuming data includes relevant period)
# Example: Filter data for a specific period (e.g., 2020-2021) and compare unemployment rates
covid_period_data = df[(df[' Date'] >= '2020-01-01') & (df[' Date'] <= '2021-12-31')]

if not covid_period_data.empty:
    avg_unemployment_covid = covid_period_data[' Estimated Unemployment Rate (%)'].mean()
    print("\nAverage Unemployment Rate during COVID-19 Period (2020-2021):", avg_unemployment_covid)
else:
    print("\nNo data found for the specified COVID-19 period.")

```

### Further analysis: Time series analysis

```python

# Group data by date and calculate the mean unemployment rate for each date
date_unemployment = df.groupby(' Date')[' Estimated Unemployment Rate (%)'].mean()

# Plot the time series
plt.figure(figsize=(12, 6))
plt.plot(date_unemployment.index, date_unemployment.values)
plt.xlabel('Date')
plt.ylabel('Average Unemployment Rate (%)')
plt.title('Time Series Analysis of Unemployment Rate')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()


# Further analysis: Correlation analysis (if other economic indicators are available)
# Assuming 'Estimated Employed' and 'Estimated Labour Participation Rate (%)' exist
# You might replace these with other relevant columns if available in your data
if ' Estimated Employed' in df.columns and ' Estimated Labour Participation Rate (%)' in df.columns:
  correlation = df[[' Estimated Unemployment Rate (%)', ' Estimated Employed', ' Estimated Labour Participation Rate (%)']].corr()
  print("\nCorrelation Matrix:\n", correlation)

  # Visualize correlation using a heatmap
  plt.figure(figsize=(8, 6))
  sns.heatmap(correlation, annot=True, cmap='coolwarm')
  plt.title('Correlation between Unemployment Rate and other Economic Indicators')
  plt.show()
else:
  print("\nMissing necessary columns for correlation analysis.")

# Further analysis: Regional comparisons
# Boxplot of unemployment rate by region
plt.figure(figsize=(12, 6))
sns.boxplot(x='Region', y=' Estimated Unemployment Rate (%)', data=df)
plt.xlabel('Region')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate by Region')
plt.xticks(rotation=45)
plt.show()
```

### Further analysis: Regional comparisons over time

```python

# Group data by region and date, then calculate the mean unemployment rate
region_date_unemployment = df.groupby(['Region', ' Date'])[' Estimated Unemployment Rate (%)'].mean().reset_index()

# Pivot the table to have regions as columns and dates as the index
region_date_unemployment_pivot = region_date_unemployment.pivot(index=' Date', columns='Region', values=' Estimated Unemployment Rate (%)')

# Plot the unemployment rates for each region over time
plt.figure(figsize=(15, 8))
for region in region_date_unemployment_pivot.columns:
    plt.plot(region_date_unemployment_pivot.index, region_date_unemployment_pivot[region], label=region)

plt.xlabel('Date')
plt.ylabel('Average Unemployment Rate (%)')
plt.title('Unemployment Rate Trends by Region Over Time')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

### Further analysis:  Focus on specific regions

```python

# Select specific regions for detailed analysis
regions_of_interest = ['Andhra Pradesh', 'Kerala', 'Delhi']
selected_data = df[df['Region'].isin(regions_of_interest)]

# Group the selected data by region and date
selected_region_date_unemployment = selected_data.groupby(['Region', ' Date'])[' Estimated Unemployment Rate (%)'].mean().reset_index()

# Pivot the table
selected_region_date_unemployment_pivot = selected_region_date_unemployment.pivot(index=' Date', columns='Region', values=' Estimated Unemployment Rate (%)')

# Plot the unemployment rates
plt.figure(figsize=(15, 8))
for region in selected_region_date_unemployment_pivot.columns:
    plt.plot(selected_region_date_unemployment_pivot.index, selected_region_date_unemployment_pivot[region], label=region)

plt.xlabel('Date')
plt.ylabel('Average Unemployment Rate (%)')
plt.title('Unemployment Rate Trends for Selected Regions Over Time')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### *Step 5: Focus on COVID-19 Impact*
